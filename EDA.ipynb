Convertirmos la columna HORA a un str de forma hh:mm

```python
def convertir_hora(hora):
    # Verificar si la hora es nula o ya está en formato estándar (contiene ":")
    if pd.isnull(hora) or ":" in str(hora):
        return hora
    else:
        # Convertir la hora a cadena y rellenar con ceros a la izquierda si es necesario
        hora_str = str(hora).zfill(4)
        # Reformatar la cadena a formato "HH:mm"
        return f"{hora_str[:2]}:{hora_str[2:]}"

# Aplicar la función a la columna 'HORA declarada'
df['HORA declarada'] = df['HORA declarada'].apply(convertir_hora)
df_1['HORA declarada'] = df_1['HORA declarada'].apply(convertir_hora)
```

Visualizamos los outliers

```python
import pandas as pd
import matplotlib.pyplot as plt

# Obtener columnas numéricas
numerical_cols = df_1.select_dtypes(include=['float64', 'int64']).columns

# Diccionario para almacenar la cantidad de outliers por columna
outliers_dict = {}

# Bucle para calcular e identificar outliers en cada columna numérica
for col in numerical_cols:
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df_1[(df_1[col] < (Q1 - 1.5 * IQR)) | (df_1[col] > (Q3 + 1.5 * IQR))].shape[0]
    outliers_dict[col] = outliers

# Convertir el diccionario a un DataFrame para facilitar la visualización
outliers_df = pd.DataFrame(list(outliers_dict.items()), columns=['Columna', 'Cantidad de Outliers'])

# Crear un gráfico de barras para visualizar la cantidad de outliers por columna
plt.figure(figsize=(10, 6))
plt.bar(outliers_df['Columna'], outliers_df['Cantidad de Outliers'])
plt.xticks(rotation=45)
plt.xlabel('Columna')
plt.ylabel('Cantidad de Outliers')
plt.title('Cantidad de Outliers por Columna')
plt.show()
```

KPI : tasa de fatalidad de la poblacion

```python
# Filtrar el DataFrame para los últimos 10 años
df_ultimos_10 = df_1[df_1['anio'] >= 2013]

# Filtrar el DataFrame para la década anterior
df_decada_anterior = df_1[(df_1['anio'] >= 2003) & (df_1['anio'] < 2013)]

# Calcular la tasa de fatalidad de la tripulación para los últimos 10 años
tasa_fatalidad_ultimos_10 = df_ultimos_10['tripulantes fallecidos'].sum() / len(df_ultimos_10)

# Calcular la tasa de fatalidad de la tripulación para la década anterior
tasa_fatalidad_decada_anterior = df_decada_anterior['tripulantes fallecidos'].sum() / len(df_decada_anterior)

# Calcular la disminución en la tasa de fatalidad
disminucion = (tasa_fatalidad_decada_anterior - tasa_fatalidad_ultimos_10) / tasa_fatalidad_decada_anterior * 100
```

```python
plt.figure(figsize=(10,6))
plt.bar(['Década Anterior', 'Últimos 10 Años'], [tasa_fatalidad_decada_anterior, tasa_fatalidad_ultimos_10])
plt.title('Tasa de Fatalidad de la Tripulación')
plt.ylabel('Tasa de Fatalidad')
plt.show()
```

Vemos que la tasa de fatalidad de la tripulacion es levemente superior en la los ultimos 10 años , que en los diez anteriores a estos , a pesar que la cantidad de accidentes a disminuido esto indica que en los accidentes que ocurrierron hubo mayor numero de muertos un promedio, quiza por el aumento de vuelo de aviones con muchos pasajeros debido al avaratamiento de los costos que se dio en la última década

```python
# Calcular la tasa de supervivencia de pasajeros para los últimos 10 años
tasa_supervivencia_pasajeros_ultimos_10 = (1 - df_ultimos_10['pasajeros fallecidos'].sum() / df_ultimos_10['pasajeros a bordo'].sum()) * 100

# Calcular la tasa de supervivencia de pasajeros para la década anterior
tasa_supervivencia_pasajeros_decada_anterior = (1 - df_decada_anterior['pasajeros fallecidos'].sum() / df_decada_anterior['pasajeros a bordo'].sum()) * 100
```

```python
plt.figure(figsize=(10,6))
plt.bar(['Década Anterior', 'Últimos 10 Años'], [tasa_supervivencia_pasajeros_decada_anterior, tasa_supervivencia_pasajeros_ultimos_10])
plt.title('Tasa de Supervivencia de Pasajeros')
plt.ylabel('Tasa de Supervivencia (%)')
plt.show()
```
